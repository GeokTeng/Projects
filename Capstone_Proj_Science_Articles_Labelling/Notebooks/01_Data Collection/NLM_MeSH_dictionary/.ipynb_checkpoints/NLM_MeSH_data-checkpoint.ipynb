# About this notebook:

- We are creating dictionaries for subsequent article labelling.
- We downloaded the MeSH terms data from the website [US National Library of Medicine (NLM)](https://www.nlm.nih.gov/databases/download/mesh.html)

**About MeSH**
- The Medical Subject Headings (MeSH) thesaurus is a controlled and hierarchically-organized vocabulary produced by the National Library of Medicine. 
- It is used for indexing, cataloging, and searching of biomedical and health-related information. 
- MeSH includes the subject headings appearing in MEDLINE/PubMed, the NLM Catalog, and other NLM databases.

```python
# Import libaries
import numpy as np 
import pandas as pd 

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

#Visualisation:
import seaborn               as sns
import matplotlib.pyplot     as plt
sns.set_theme(style="whitegrid")
```

# Preparing MeSH terms dataframe

```python
# Read the data from the BIN file
with open('mtrees2022.bin', 'r') as f:
    lines = f.readlines()

# Create a list of dictionaries with the Term and term_full_ID for each line
data = []
for line in lines:
    term, term_full_ID = line.strip().split(';')
    data.append({'Term': term, 'term_full_ID': term_full_ID})

print(len(data))
```

```python
# Create a Pandas dataframe from the list of dictionaries
df = pd.DataFrame(data)

# Print the dataframe
print(df)
```

```python
df['Root_term'] = df['term_full_ID'].apply(lambda x: x[0])
df.head()
```

```python
disease_df = df.copy()
```

```python
disease_df['Root_term'].unique()
disease_df['Root_term'].nunique()
```

```python
disease_df = disease_df.loc[disease_df['Root_term']=='C',:]
disease_df
```

```python
disease_df['Pri_Disease_ID'] = disease_df['term_full_ID'].apply(lambda x: x[:3])
disease_df['Sec_Disease_ID'] = disease_df['term_full_ID'].apply(lambda x: x[:7])
```

```python
disease_df['Sec_Disease_ID']
```

```python
disease_df['Pri_Disease_ID'].unique()
disease_df['Pri_Disease_ID'].nunique()
```

```python
disease_df.drop(columns="Root_term",inplace=True)
disease_df
```

# Analysing the Mesh Terms (EDA)

## Primary disease ID & Associate Primary disease types

```python
# key = Primary disease ID
# value = Associate Primary disease type
_df = disease_df.loc[(disease_df["term_full_ID"].str.len())<4,:]
Pri_disease_dict = dict(zip(_df['term_full_ID'], _df['Term']))
Pri_disease_dict
```

```python
print(f"Length of Pri_disease_dict: {len(Pri_disease_dict)}")
```

```python
disease_df['Pri_disease_term'] = disease_df['Pri_Disease_ID'].map(Pri_disease_dict)
disease_df
```

### Diseases under each Pri_Disease_ID

```python
_s = disease_df.groupby('Pri_Disease_ID').size()
print("Number of diseases under each Pri_Disease_ID")
_s
print(f"\nStatistical description of Primary term ID: \n{_s.describe()}")
```

```python
sns.histplot(_s, bins=20)
plt.title("Number of diseases under each Pri_Disease_ID");
```

## Secondary disease ID & Associate Secondary disease types

```python
length = disease_df["term_full_ID"].str.len()

disease_df.loc[(length>3)&(length<8),:]
```

```python
#dictionary refelecting each key and the associate disease as value
_df = disease_df.loc[(length>3)&(length<8),:]
Sec_disease_dict = dict(zip(_df['Sec_Disease_ID'], _df['Term']))
Sec_disease_dict
```

```python
print(f"Length of Sec_disease_dict: {len(Sec_disease_dict)}")
```

```python
disease_df['Sec_disease_term'] = disease_df['Sec_Disease_ID'].map(Sec_disease_dict)
disease_df
```

```python
disease_df[disease_df["Term"]=='22q11 Deletion Syndrome']
```

### Diseases under each Sec_Disease_ID

```python
print("Number of diseases under each Sec_Disease_ID:")
disease_df.groupby('Sec_Disease_ID').size()
_s = disease_df.groupby('Sec_Disease_ID').size()
print(f"\nStatistical description of Secondary term ID: \n{_s.describe()}")
```

```python
sns.histplot(_s)
plt.title("Number of diseases under each Sec_Disease_ID");
```

### Secondary terms with multiple Sec_disease ID

```python
# Create a new dictionary to store the keys grouped by their values
grouped_dict = {}

# Iterate over the keys and values of the original dictionary
for key, value in Sec_disease_dict.items():
    
    # Use setdefault method to add value to the list of keys with the same value
    grouped_dict.setdefault(value, []).append(key)
```

```python
# Iterate over the keys and values of the grouped dictionary
print("Secondary terms with multiple Sec_disease ID:\n\n")
for key, value in grouped_dict.items():
    
    # Print the keys with the same value
    if len(value) > 1:
        print("Disease:", key, "\n -----> ID:", value, "\n ")
```

## Creating the important dictionary

Each unique term forms the 'key', with its value containins a list of secondary ids

### Sec_term

```python
sec_term_dict = disease_df.groupby('Term')['Sec_Disease_ID'].apply(list).to_dict()
len(sec_term_dict)
```

```python
disease_df['Term'].nunique()
```

```python
sec_term_dict
```

```python
#remove duplicate values in each key
for key, value in sec_term_dict.items():
    sec_term_dict[key]=set(value)
```

### Pri_term

```python
pri_term_dict = disease_df.groupby('Term')['Pri_Disease_ID'].apply(list).to_dict()
len(pri_term_dict)
```

```python
pri_term_dict
```

```python
#remove duplicate values in each key
for key, value in pri_term_dict.items():
    pri_term_dict[key]=set(value)
```

# Analysing the Pri term & Sec terms

```python
terms_df = disease_df.loc[:,['Pri_Disease_ID', 'Sec_Disease_ID']]
terms_df.drop_duplicates(inplace=True)
terms_df.drop(terms_df[terms_df['Pri_Disease_ID'] == terms_df['Sec_Disease_ID']].index, inplace=True)
```

```python
terms_df
```

```python
print("Number of Secondary ID under each Primary ID:")
_s = terms_df.groupby('Pri_Disease_ID')['Sec_Disease_ID'].size()
_s
```

```python
def annotation_fx(my_barchart):
    for each_bar in my_barchart.patches:
        plt.annotate(format(each_bar.get_height(), '.0f'),
                       (each_bar.get_x() + each_bar.get_width() / 2,
                        each_bar.get_height()), ha='center', va='center',
                       size=10, xytext=(0, 8),
                       textcoords='offset points')
```

```python
fig = plt.figure(figsize=(10,8))

_s.sort_values().plot(kind='barh')

# add annotations to the bars
ax = _s.sort_values().plot(kind='barh')
for i, v in enumerate(_s.sort_values()):
    ax.text(v + 0.5, i - 0.1, str(v), color='black')

# set the Y-tick labels using the label_dict
ax.set_yticklabels([Pri_disease_dict[x] for x in _s.index])
plt.title("Number of Secondary ID under each Primary ID",fontsize=16)
plt.ylabel("Primary ID");
```
