## `Period`

```python
# Define the periods
periods = {
    (1990, 1995): '1990-1995',
    (1996, 2000): '1996-2000',
    (2001, 2005): '2001-2005',
    (2006, 2010): '2006-2010',
    (2011, 2015): '2011-2015',
    (2016, 2016): '2016'
}

# Function to map years to periods
def map_to_period(year):
    for period_range, period_value in periods.items():
        start, end = period_range
        if start <= year <= end:
            return period_value
```

```python
for data in dataframe_ls:
    print(f"\nDataframe{data}")
    df = globals()[data].copy()
    df['Period'] = df['Year'].apply(map_to_period)
    
    globals()[data] = df
    globals()[data].head()
```

# Malaria Death By Country and its Continent

```python
#Label the countries continental code
gc = GeonamesCache()

# Access data for specific entities
gc_countries = gc.get_countries()

# List of country codes in GeonamesCache
gc_country_codes = [value['iso3'] for key, value in gc_countries.items()]
# List of countries respective continental codes GeonamesCache
gc_country_cont_codes = [value['continentcode'] for key, value in gc_countries.items()]

gc_country_cont_dict = {k:v for k,v in zip(gc_country_codes,gc_country_cont_codes)}
```

```python
UK_countries = {'England','Scotland','Wales','Northern Ireland'}

country_df_ls=[]
for data in dataframe_ls:
    subdata = "country_"+data
    country_df_ls.append(subdata)
    
    #creating subset data with only countries
    df = globals()[data].copy()
    subdf = df.loc[df['Entity_type']=='country',:].copy()
    
    #Labelling countries according to their continent code
    subdf['Continent'] = subdf['Code'].map(gc_country_cont_dict)
    subdf.loc[subdf['Entity'].isin(UK_countries),'Continent'] = 'EU'
    
    globals()[subdata] = subdf
    
    print(f"Dataframe:{subdata}")
    globals()[subdata].head()
```
