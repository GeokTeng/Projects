## Function Creation
### Dataframe Exploratory Funtions

```python
#Showing missing, duplicates, shape, dtypes
def df_summary(df):
    print(f"Shape(col,rows): {df.shape}")
    print(f"Number of duplicates: {df.duplicated().sum()}")
    print('---'*20)
    print(f'Number of each unqiue datatypes:\n{df.dtypes.value_counts()}')
    print('---'*20)
    print("Columns with missing values:")
    isnull_df = pd.DataFrame(df.isnull().sum()).reset_index()
    isnull_df.columns = ['col','num_nulls']
    isnull_df['perc_null'] = ((isnull_df['num_nulls'])/(len(df))).round(2)
    _df_ = isnull_df[isnull_df['num_nulls']>0]
    if _df_.empty:
        print("--No Missing Data--")
    else:
        print(_df_)
```

### Visualisation Functions

```python
#Visualisation

#Correlation heatmap
def plot_correlation(df):
    corr = df.corr(numeric_only=True).round(2)
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    sns.heatmap(corr,
                cmap = sns.diverging_palette(220, 20, n = 8),
                annot = True,
                vmin = -1,
                vmax = 1,
                linewidths = 0.5,
                annot_kws={"size":5},
                cbar_kws={"shrink": 0.5},
                mask = mask
                   )
```

### Other Functions

```python
#return 25th, 50th, 75th percentile 

def q25(x):
    return x.quantile(0.25)
def q50(x):
    return x.quantile(0.5)
def q75(x):
    return x.quantile(0.75)
```

```python
#return dataframe showing outliers only, if value is not outlier will show as Nan
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

## **Observation & Remarks:**


|| Observation:| Remarks:|
|---|---|---|
|Longitude & Latitudes| Facilities Longitude and Latitude would reasonably have strongly correlate with that of the resale unit's Longitude and Latitude.|Thus, further analysis shall drop columns **containing Facilities Longitude and Latitude**|
|HDA Age| HDB Age, year_completed and lease_comence_date share collinearity.|As such, further analysis shall drop columns **'year_completed'** and **'lease_comence_date'.**|
|Feature with significant correlation with resale price: Floor area|Among all the numeric features, only floor area has significant correlation with resale price (r=0.65)|It make sense for larger house with fetch higher pricees|

# EDA Part 8: Statsmodel - OLS<a class="anchor" id="ID8"></a>

```python
import statsmodels.api as sm
```

```python
#create a subset dataframe with only numerical columns
_ls = df.select_dtypes(include=['int','float']).columns.tolist()
_ls.remove('id')
_df = df.loc[:,_ls]
df_summary(_df)
```

```python
#determine which feature is significant/has statistical correlation via OLS
_ls.remove('resale_price')
X = _df[_ls]
y = _df['resale_price']

X = sm.add_constant(X, prepend=True)
results = sm.OLS(y, X).fit()
```

```python
coef_table = pd.DataFrame(results.summary().tables[1].data)

# Set column names
coef_table.columns = coef_table.iloc[0]
coef_table = coef_table[1:]

# Filter for features with P-value greater than 0.05
significant_features = coef_table[coef_table['P>|t|'].astype(float) > 0.05]
significant_features
```

### **Observation & Remarks**

- These 5 features appear to have no statistical signficant effect to resale prices (P>|t| > 5%)

**Features retained for model training:**
- Categorical features
- Numerical features excluding ID and the above 6 columns that does not appear to have significant relation with resale price:
    - 'id'
    - 'Have_market_hawker'
    - 'multigen_sold'
    - '3room_rental'
    - 'Mall_Nearest_Distance'
    - 'Hawker_Within_500m'
    - 'bus_stop_nearest_distance	'

```python
col_to_excl = significant_features.iloc[:,0].tolist()
col_to_excl.append('id')
col_to_excl
```

```python
df.shape
```

```python
new_df = df.copy()
new_df.drop(columns = col_to_excl, inplace=True)
new_df.shape
```
